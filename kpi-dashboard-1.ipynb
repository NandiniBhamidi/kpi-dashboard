```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(42)  # For reproducibility

# Parameters
n_jobs = 1000
min_duration = 5  # in minutes
max_duration = 120  # in minutes
min_data_processed = 1  # in GB
max_data_processed = 500  # in GB
min_compute_hours = 0.5
max_compute_hours = 10
cost_per_compute_hour = 0.25  # in USD

# Generate Data
job_ids = range(1, n_jobs + 1)
start_times = [datetime.now() - timedelta(days=np.random.randint(1, 60)) for _ in range(n_jobs)]
durations = np.random.randint(min_duration, max_duration, n_jobs)
#end_times = [start_times[i] + timedelta(minutes=durations[i]) for i in range(n_jobs)]
data_processed = np.random.uniform(min_data_processed, max_data_processed, n_jobs).round(2)
job_statuses = np.random.choice(['Success', 'Failed'], n_jobs, p=[0.95, 0.05])
compute_hours_used = np.random.uniform(min_compute_hours, max_compute_hours, n_jobs).round(2)
costs = compute_hours_used * cost_per_compute_hour

# Create DataFrame
etl_jobs_df = pd.DataFrame({
    'Job ID': job_ids,
    'Start Time': start_times,
 #   'End Time': end_times,
    'Duration (min)': durations,
    'Data Processed (GB)': data_processed,
    'Job Status': job_statuses,
    'Compute Hours Used': compute_hours_used,
    'Cost (USD)': costs
})

etl_jobs_df.head()  # Display the first few rows
```
